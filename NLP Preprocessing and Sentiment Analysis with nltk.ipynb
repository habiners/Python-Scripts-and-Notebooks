```python
df['text'].replace(to_replace="([^\w\s]|_|[^\x00-\x7F])+", value="", regex=True, inplace=True)
```

```python
from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer

ps = PorterStemmer() 
sbstemmer = SnowballStemmer("english") # Better daw ni
```

```python
textColumn = list()
for index in range(df['text'].size):
    tokenizedRow = df['text'][index].split(' ')
    stemmed = ""
    for word in tokenizedRow:
        stemmed += (sbstemmer.stem(word) + " ")
    textColumn.append(stemmed[:-1])
df['text'] = textColumn
```

```python
df.to_csv('038 Coachella-E5-2-DFE Preprocessed.csv', index = False)
```
