```python
df = pd.read_csv('mushrooms.csv') # Source: https://www.kaggle.com/uciml/mushroom-classification
df.head()
```

```python
for col in df.columns:
    if col != "class":
        sns.catplot(x=col, col="class", data=df, kind="count", palette =['red','green'], height=3)
```

```python
for col in df.columns:
    df[col] = df[col].astype('category').cat.codes
```

```python
df.head()
```

```python
df.hist(figsize = (15,15))
```

```python
print(df['veil-type'].value_counts())
del df['veil-type']
df.head()
```

```python
x_features = df.drop('class', axis=1)
y_to_predict = df['class']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_features = pd.DataFrame(scaler.fit_transform(x_features), columns=x_features.columns)
x_features
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(), annot=True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_features, y_to_predict, test_size=0.25, random_state = 42)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
best_knn_model = None
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(x_train, y_train)
    
    if (not best_knn_model or best_knn_model.score(x_test,y_test) <= knn.score(x_test,y_test)):
        best_knn_model = knn
    
    print(f'KNeighborsClassifier: k = {knn.n_neighbors} , Accuracy is: {knn.score(x_test,y_test)}')
print(f'Best KNeighborsClassifier: k = {best_knn_model.n_neighbors} , Accuracy is: {best_knn_model.score(x_test,y_test)}')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, knn.predict(x_test)))
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(knn, x_test, y_test, display_labels= ['Edible', 'Poisonous'])
plt.title(f'Confusion Matrix KNN with k = {knn.n_neighbors}')
plt.show()

from sklearn import metrics
print(f'Confusion matrix Accuracy is: {metrics.accuracy_score(y_test, knn.predict(x_test))}')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, best_knn_model.predict(x_test)))
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(best_knn_model, x_test, y_test, display_labels= ['Edible', 'Poisonous'])
plt.title(f'Confusion Matrix KNN with k = {best_knn_model.n_neighbors}')
plt.show()

from sklearn import metrics
print(f'Confusion matrix Accuracy is: {metrics.accuracy_score(y_test, best_knn_model.predict(x_test))}')
```

```python
print("Training Set: ")
print(f"Dataset: \t{y_train.values[0:30]}")
print(f"Prediction: \t{best_knn_model.predict(x_train.values[0:30])}")

print("\nTest Set: ")
print(f"Dataset: \t{y_test.values[0:30]}")
print(f"Prediction: \t{best_knn_model.predict(x_test.values[0:30])}")
```
